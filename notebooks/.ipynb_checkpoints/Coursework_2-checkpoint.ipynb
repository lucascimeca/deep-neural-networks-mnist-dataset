## Experiment 1

### Set up

```python
# SET UP
import numpy as np
import logging
import time
from mlp.data_providers import MNISTDataProvider

import matplotlib.pyplot as plt
from mlp.layers import AffineLayer, SoftmaxLayer, SigmoidLayer, ReluLayer, TanhLayer
from mlp.errors import CrossEntropySoftmaxError
from mlp.models import MultipleLayerModel
from mlp.initialisers import ConstantInit, GlorotUniformInit, UniformInit
from mlp.optimisers import Optimiser
from mlp.learning_rules import GradientDescentLearningRule, RMSPropLearningRule, MomentumLearningRule
from mlp.schedulers import ConstantLearningRateScheduler, ExponentialLearningRateScheduler
%matplotlib inline
plt.style.use('ggplot')


# Seed a random number generator
seed = 10102016 
rng = np.random.RandomState(seed)

# Set up a logger object to print info about the training run to stdout
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.handlers = [logging.StreamHandler()]

# Create data provider objects for the MNIST data set
train_data = MNISTDataProvider('train', batch_size=50, rng=rng)
valid_data = MNISTDataProvider('valid', batch_size=50, rng=rng)

def train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval, schedulers=[]):
    
    data_monitors={'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
    
    optimiser = Optimiser(
        model, error, learning_rule, train_data, valid_data, data_monitors, schedulers
    )
    
    stats, keys, run_time = optimiser.train(num_epochs=num_epochs, stats_interval=stats_interval)
    
    return stats, keys, run_time


def plot_stats(x, y, x_label='x', y_label='y', name='name'):
    fig_size = (6, 3) 
    fig = plt.figure(figsize=fig_size)
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, linewidth=2, label='$'+name+'$')
    ax.set_xlabel('$'+x_label+'$', fontsize=12)
    ax.set_ylabel('$'+y_label+'$', fontsize=12)
    ax.grid('on')
    ax.legend(loc='best', fontsize=11) 
    fig.tight_layout() 
    fig.savefig(name+'.pdf') 
```

### Saturation


```python
input_dim, output_dim, hidden_dim = 784, 10, 100

# Set training run hyperparameters
batch_size = 50  # number of data points in a batch
init_scale = 0.1  # scale for random parameter initialisation
learning_rate = 0.07  # learning rate for gradient descent
num_epochs = 100  # number of training epochs to perform
stats_interval = 5  # epoch interval between recording and printing stats

lows = [.000001, .001, .01, .1]
highs = [.00001, .01, .1, .3]

for i in range(len(lows)):

    weights_init = UniformInit(rng=rng, low=lows[i], high=highs[i])
    biases_init = ConstantInit(0.)
    learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)
    
    #-------------------------SIGMOID--------------------------------------
    train_data.reset()
    valid_data.reset()

    model = MultipleLayerModel([
        AffineLayer(input_dim, hidden_dim, weights_init, biases_init), 
        SigmoidLayer(),
        AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
    ])

    error = CrossEntropySoftmaxError()

    # Use a basic gradient descent learning rule

    stats, keys, _, = train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)
    softmax_error = stats[1:, keys['error(train)']]

    #-------------------------RELU--------------------------------------
    train_data.reset()
    valid_data.reset()

    model = MultipleLayerModel([
        AffineLayer(input_dim, hidden_dim, weights_init, biases_init), 
        ReluLayer(),
        AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
    ])

    error = CrossEntropySoftmaxError()

    stats, keys, _, = train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)
    relu_error = stats[1:, keys['error(train)']]

    #-------------------------TANH--------------------------------------
    train_data.reset()
    valid_data.reset()

    model = MultipleLayerModel([
        AffineLayer(input_dim, hidden_dim, weights_init, biases_init), 
        TanhLayer(),
        AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
    ])

    error = CrossEntropySoftmaxError()

    stats, keys, _, = train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)
    tanh_error = stats[1:, keys['error(train)']]

    fig_size = (6, 3) 
    fig = plt.figure(figsize=fig_size)
    ax = fig.add_subplot(1, 1, 1)

    ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                  softmax_error, label='Sigmoid')
    ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                  relu_error, label='ReLU')
    ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                  tanh_error, label='Tanh')

    ax.set_xlabel('$epoch$', fontsize=12)
    ax.set_ylabel('$training\_error$', fontsize=12)
    ax.grid('on')
    ax.legend(loc='best', fontsize=11) 
    fig.tight_layout() 
    fig.savefig('sig_relu_tan_gain_'+str(i)+'.pdf') 
```

## Experiment 2

```python
# SET UP
import numpy as np
import logging
import time
from mlp.data_providers import MNISTDataProvider

import matplotlib.pyplot as plt
from mlp.layers import AffineLayer, SigmoidLayer, BatchNormLayer
from mlp.errors import CrossEntropySoftmaxError
from mlp.models import MultipleLayerModel
from mlp.initialisers import ConstantInit, UniformInit, GlorotUniformInit
from mlp.optimisers import Optimiser
from mlp.learning_rules import RMSPropLearningRule, GradientDescentLearningRule, MomentumLearningRule
from mlp.schedulers import ConstantLearningRateScheduler, ExponentialLearningRateScheduler
%matplotlib inline
plt.style.use('ggplot')


# Seed a random number generator
seed = 10102016 
rng = np.random.RandomState(seed)

# Set up a logger object to print info about the training run to stdout
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.handlers = [logging.StreamHandler()]

# Create data provider objects for the MNIST data set
train_data = MNISTDataProvider('train', batch_size=50, rng=rng)
valid_data = MNISTDataProvider('valid', batch_size=50, rng=rng)

def train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval, schedulers=[]):
    
    data_monitors={'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
    
    optimiser = Optimiser(
        model, error, learning_rule, train_data, valid_data, data_monitors, schedulers
    )
    
    stats, keys, run_time = optimiser.train(num_epochs=num_epochs, stats_interval=stats_interval)
    
    return stats, keys, run_time

def plot_stats(x, y, x_label='x', y_label='y', name='name'):
    fig_size = (6, 3) 
    fig = plt.figure(figsize=fig_size)
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, linewidth=2, label='$'+name+'$')
    ax.set_xlabel('$'+x_label+'$', fontsize=12)
    ax.set_ylabel('$'+y_label+'$', fontsize=12)
    ax.grid('on')
    ax.legend(loc='best', fontsize=11) 
    fig.tight_layout() 
    fig.savefig(name+'.pdf') 
```

### initialization robustness

```python
# input_dim, output_dim, hidden_dim = 784, 10, 100

# Set training run hyperparameters
batch_size = 50  # number of data points in a batch
init_scale = 0.1  # scale for random parameter initialisation
learning_rate = .1  # learning rate for gradient descent
num_epochs = 100  # number of training epochs to perform
stats_interval = 5  # epoch interval between recording and printing stats

weights_init = UniformInit(rng=rng, low=0.1, high=0.3)
biases_init = ConstantInit(0.)
learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)

# #-------------------------REGULAR--------------------------------------
train_data.reset()
valid_data.reset()

model = MultipleLayerModel([
    AffineLayer(input_dim, hidden_dim, weights_init, biases_init),
    SigmoidLayer(),
    AffineLayer(hidden_dim, hidden_dim, weights_init, biases_init),
    SigmoidLayer(),
    AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
])

error = CrossEntropySoftmaxError()

# Use a basic gradient descent learning rule

stats, keys, regular_time, = train_model(
    model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)

fig_size = (6, 3) 
fig = plt.figure(figsize=fig_size)
ax = fig.add_subplot(1, 1, 1)

# regular_error = stats[1:, keys['error(train)']]
# regular_valid = stats[1:, keys['error(valid)']]

ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
              regular_error, label='non-batch training error')
ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
              regular_valid, label='non-batch validation error')

ax.set_xlabel('$epoch$', fontsize=12)
ax.set_ylabel('$error$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 

#-------------------------BATCH NORMALIZATION--------------------------------------

train_data.reset()
valid_data.reset()

model = MultipleLayerModel([
    AffineLayer(input_dim, hidden_dim, weights_init, biases_init), 
    BatchNormLayer(input_dim=hidden_dim),
    SigmoidLayer(),
    AffineLayer(hidden_dim, hidden_dim, weights_init, biases_init),
    SigmoidLayer(),
    AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
])

error = CrossEntropySoftmaxError()

# Use a basic gradient descent learning rule

stats, keys, batch_norm_time, = train_model(
    model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)

ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
              stats[1:, keys['error(train)']], label='batch training error')
ax.plot(np.arange(1, stats.shape[0]) * stats_interval, 
              stats[1:, keys['error(valid)']], label='batch validation error')

ax.set_xlabel('$epoch$', fontsize=12)
ax.set_ylabel('$error$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 
fig.savefig('batch_test2.pdf') 
```

### classification performance

#### eta search


```python
# find best hyperparameter settings
input_dim, output_dim, hidden_dim = 784, 10, 100

# Set training run hyperparameters
batch_size = 50  # number of data points in a batch
init_scale = 0.1  # scale for random parameter initialisation
learning_rate = .1  # learning rate for gradient descent
num_epochs = 100  # number of training epochs to perform
stats_interval = 5  # epoch interval between recording and printing stats


weights_init = GlorotUniformInit(rng=rng)
biases_init = ConstantInit(0.)

learning_rates = np.linspace(0.001, 5., 10)
valid_errors = []
min_rule = (0, 100.)
for learning_rate in learning_rates:
    learning_rule = MomentumLearningRule(learning_rate=learning_rate,  mom_coeff=0.9)

    train_data.reset()
    valid_data.reset()

    model = MultipleLayerModel([
        AffineLayer(input_dim, hidden_dim, weights_init, biases_init), 
        BatchNormLayer(input_dim=hidden_dim),
        SigmoidLayer(),
        AffineLayer(hidden_dim, hidden_dim, weights_init, biases_init),
        SigmoidLayer(),
        AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
    ])


    error = CrossEntropySoftmaxError()

    # Use a basic gradient descent learning rule

    stats, keys, batch_norm_time, = train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)
    
    minimum = min(stats[1:, keys['error(valid)']])
    if minimum < min_rule[1]:
        min_rule = (learning_rate, minimum)
    valid_errors += [minimum]

print(min_rule)

fig_size = (6, 3) 
fig = plt.figure(figsize=fig_size)
ax = fig.add_subplot(1, 1, 1)

ax.plot(learning_rates, valid_errors, label='learning rate search')

ax.set_xlabel('$learning\_rate (\eta)$', fontsize=12)
ax.set_ylabel('$minimum\_validation\_error$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 
fig.savefig('batch_test_hyper_search.pdf') 
```

## Experiment 3

```python
# SET UP
import numpy as np
import logging
import time
from mlp.data_providers import MNISTDataProvider

import matplotlib.pyplot as plt
from mlp.layers import (
    AffineLayer, 
    SoftmaxLayer, 
    ConvolutionalLayer, 
    MaxPoolingLayer, 
    ReshapeLayer, 
    LayerWithParameters, 
    SigmoidLayer, 
    BatchNormLayer,
    ReluLayer
)
from mlp.errors import CrossEntropySoftmaxError
from mlp.models import MultipleLayerModel
from mlp.initialisers import ConstantInit, GlorotUniformInit
import mlp.initialisers as init
from mlp.optimisers import Optimiser
from mlp.learning_rules import GradientDescentLearningRule, RMSPropLearningRule, MomentumLearningRule
from mlp.schedulers import ConstantLearningRateScheduler, ExponentialLearningRateScheduler
%matplotlib inline
plt.style.use('ggplot')


# Seed a random number generator
seed = 10102016 
rng = np.random.RandomState(seed)

# Set up a logger object to print info about the training run to stdout
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.handlers = [logging.StreamHandler()]

# Create data provider objects for the MNIST data set
train_data = MNISTDataProvider('train', batch_size=50, rng=rng)
valid_data = MNISTDataProvider('valid', batch_size=50, rng=rng)

short_train_data = MNISTDataProvider('train', batch_size=50, rng=rng, max_inputs_num=5000)
short_val_data = MNISTDataProvider('valid', batch_size=50, rng=rng, max_inputs_num=2000)


def train_model(
        model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval, schedulers=[]):
    
    data_monitors={'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
    
    optimiser = Optimiser(
        model, error, learning_rule, train_data, valid_data, data_monitors, schedulers
    )
    
    stats, keys, run_time = optimiser.train(num_epochs=num_epochs, stats_interval=stats_interval)
    
    return stats, keys, run_time


def plot_stats(x, y, x_label='x', y_label='y', name='name'):
    fig_size = (6, 3) 
    fig = plt.figure(figsize=fig_size)
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, linewidth=2, label='$'+name+'$')
    ax.set_xlabel('$'+x_label+'$', fontsize=12)
    ax.set_ylabel('$'+y_label+'$', fontsize=12)
    ax.grid('on')
    ax.legend(loc='best', fontsize=11) 
    fig.tight_layout() 
    fig.savefig(name+'.pdf') 
```

### Convolutional Neural Network

```python
input_dim, output_dim, hidden_dim = 784, 10, 100

# Set training run hyperparameters
batch_size = 50  # number of data points in a batch
init_scale = 0.1  # scale for random parameter initialisation
learning_rate = .1  # learning rate for gradient descent
num_epochs = 100  # number of training epochs to perform
stats_interval = 5 # epoch interval between recording and printing stats


weights_init = GlorotUniformInit(rng=rng)
biases_init = ConstantInit(0.)
learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)


#-------------------------CONV NET--------------------------------------
train_data.reset()
valid_data.reset()

model = MultipleLayerModel([
    ReshapeLayer(output_shape=(1, 28, 28)),
    ConvolutionalLayer(1, 24, 28, 28, 5, 5),
    MaxPoolingLayer(),
    ReluLayer(),
    ReshapeLayer(),
    AffineLayer(6912, hidden_dim, weights_init, biases_init), 
    ReluLayer(),
    AffineLayer(hidden_dim, output_dim, weights_init, biases_init), 
])

error = CrossEntropySoftmaxError()

conv_net_stats, conv_net_keys, conv_net_time, = train_model(
    model, error, learning_rule, train_data, valid_data, num_epochs, stats_interval)


fig_size = (6, 3) 
fig = plt.figure(figsize=fig_size)
ax = fig.add_subplot(1, 1, 1)

conv_net_train_error = conv_net_stats[1:, conv_net_keys['error(train)']]
conv_net_val_error = conv_net_stats[1:, conv_net_keys['error(valid)']]

ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_train_error, label='training error')
ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_val_error, label='validation error')

ax.set_xlabel('$epoch$', fontsize=12)
ax.set_ylabel('$error$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 
fig.savefig('CNN1.pdf') 

print('------ ConvNet-----')
print('Time elapsed: '+str(conv_net_time))
print('Min trainin error: '+str(min(conv_net_train_error)))
print('Min val error: '+str(min(conv_net_val_error)))
```

```python
fig_size = (6, 3) 
fig = plt.figure(figsize=fig_size)
ax = fig.add_subplot(1, 1, 1)

conv_net_train_acc = conv_net_stats[1:, conv_net_keys['acc(train)']]
conv_net_val_acc = conv_net_stats[1:, conv_net_keys['acc(valid)']]

ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_train_acc, label='training error')
ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_val_acc, label='validation error')

ax.set_xlabel('$epoch$', fontsize=12)
ax.set_ylabel('$accuracy$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 
fig.savefig('CNN1_acc.pdf') 
```

### Convolutional Neural Network vs CNN + Batch

```python
input_dim, output_dim, hidden_dim = 784, 10, 100

# Set training run hyperparameters
batch_size = 50  # number of data points in a batch
init_scale = 0.1  # scale for random parameter initialisation
num_epochs = 50  # number of training epochs to perform
stats_interval = 5 # epoch interval between recording and printing stats


weights_init = GlorotUniformInit(rng=rng)
biases_init = ConstantInit(0.)


learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)

#-------------------------CONV NET --------------------------------------

train_data.reset()
valid_data.reset()

model = MultipleLayerModel([
    ReshapeLayer(output_shape=(1, 28, 28)),
    ConvolutionalLayer(1, 16, 28, 28, 5, 5),
    MaxPoolingLayer(),
    ReluLayer(),
    ReshapeLayer(),
    AffineLayer(4608, hidden_dim, weights_init, biases_init), 
    ReluLayer(),
    AffineLayer(hidden_dim, output_dim, weights_init, biases_init), 
])

error = CrossEntropySoftmaxError()

conv_net_stats, conv_net_keys, conv_net_time, = train_model(
    model, error, learning_rule, short_train_data, short_val_data, num_epochs, stats_interval)

fig_size = (6, 3) 
fig = plt.figure(figsize=fig_size)
ax = fig.add_subplot(1, 1, 1)

conv_net_train_error = conv_net_stats[1:, conv_net_keys['error(train)']]
conv_net_val_error = conv_net_stats[1:, conv_net_keys['error(valid)']]

ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_train_error, label='non-batch training error')
ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_val_error, label='non-batch validation error')

learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)
#-------------------------CONV NET + batch layer--------------------------------------
train_data.reset()
valid_data.reset()

model = MultipleLayerModel([
    AffineLayer(input_dim, input_dim, weights_init, biases_init), 
    BatchNormLayer(input_dim=input_dim),
    ReshapeLayer(output_shape=(1, 28, 28)),
    ConvolutionalLayer(1, 16, 28, 28, 5, 5),
    MaxPoolingLayer(),
    ReluLayer(),
    ReshapeLayer(),
    AffineLayer(4608, hidden_dim, weights_init, biases_init),
    ReluLayer(),
    AffineLayer(hidden_dim, output_dim, weights_init, biases_init), 
        
])

error = CrossEntropySoftmaxError()

conv_net_stats, conv_net_keys, conv_net_time2, = train_model(
    model, error, learning_rule, short_train_data, short_val_data, num_epochs, stats_interval)

conv_net_train_error2 = conv_net_stats[1:, conv_net_keys['error(train)']]
conv_net_val_error2 = conv_net_stats[1:, conv_net_keys['error(valid)']]

ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_train_error2, label='batch training error')
ax.plot(np.arange(1, conv_net_stats.shape[0]) * stats_interval, 
              conv_net_val_error2, label='batch validation error')

ax.set_xlabel('$epoch$', fontsize=12)
ax.set_ylabel('$error$', fontsize=12)
ax.grid('on')
ax.legend(loc='best', fontsize=11) 
fig.tight_layout() 
fig.savefig('batch_vs_non-batch_CNN-50epochs2.pdf') 

print('------ ConvNet-----')
print('Time elapsed: '+str(conv_net_time))
print('Min trainin error: '+str(min(conv_net_train_error2)))
print('Min val error: '+str(conv_net_val_error))
print('------ ConvNet + batch-----')
print('Time elapsed: '+str(conv_net_time2))
print('Min trainin error: '+str(min(conv_net_train_error2)))
print('Min val error: '+str(min(conv_net_val_error2)))
```
